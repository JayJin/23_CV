# Numpy image

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

path = './imgs/cat.jpg'

image_pil = Image.open(path)
image = np.array(image_pil)         #   

plt.imshow(image)
plt.show
```

```python
image.shape         # (H(height), W(width), C(channel))
```

```python
# histogram
plt.hist(image.reshape(-1), 256, [0, 256])
plt.show
```

# Numpy image ROI(Region of interest)
 - Direct change a pixel : image[350, 100] = [0,0,0]
 - roi = image[y1:y2, x1:x2] (x1 =< x2, y1 =< y2)
 - Masking, Draw    

```python
eye = image[200:245, 165:240].copy()
image[200:245, 230:305] = eye

plt.imshow(image)
plt.show()
```
